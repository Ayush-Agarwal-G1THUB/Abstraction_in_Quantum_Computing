# qml.reset()

```python
import pennylane as qml

dev = qml.device("default.qubit", wires=2)

@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=0)
    m = qml.measure(wires=0, reset=True)
    
    return qml.probs(wires=0)

print(circuit()) # Should return [0, 1] because it was reset to 0 then flipped to 1.
```

reset=True sets the qubit to |0>, not to the state it was before the measurement

Documentation : https://docs.pennylane.ai/en/stable/code/api/pennylane.measure.html

So if we are trying to check for predicate fulfillment, we would not use qml.measure(), we should get the raw numpy data from qml.state() and pass that to the predicate function. The function will then treat it as a numpy array and use the mathematics to check whether the predicate is being violated or not
